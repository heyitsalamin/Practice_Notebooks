# Multi class classification of Brain Tumor

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_dir = 'dataset/brain_tumor/'
```

```python
train_generator = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255,
    zoom_range=0.2,
    horizontal_flip=True,
    rotation_range=20,
    width_shift_range=0.2,
    fill_mode='nearest',
    height_shift_range=0.1,
    shear_range=0.2,
    brightness_range=None,
    validation_split=0.2
)
val_generator = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)
```

```python
train_data = train_generator.flow_from_directory(
    data_dir,
    target_size=(224,224),
    batch_size=32,
    shuffle=True,
    class_mode='categorical',
    subset='training'
)

val_data = val_generator.flow_from_directory(
    data_dir,
    target_size=(224,224),
    batch_size=32,
    shuffle=False,
    class_mode='categorical',
    subset='validation'
)
```

```python
import os
LABELS = [f.name for f in os.scandir(data_dir) if f.is_dir()]
```

# Creating a Model

```python
from tensorflow.keras.layers import Dense, MaxPooling2D, Conv2D, Dropout, BatchNormalization, Input, Flatten, GlobalAveragePooling2D
from tensorflow.keras.models import Model

input_layer = Input(shape=(224,224,3))
x = Conv2D(32, 3, activation = 'relu',padding = 'same')(input_layer)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = MaxPooling2D(2,2)(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = MaxPooling2D(2,2)(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = MaxPooling2D(2,2)(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = MaxPooling2D(2,2)(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = MaxPooling2D(2,2)(x)
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation = 'relu')(x)
x = Dropout(0.3)(x)
output_layer = Dense(4, activation = 'softmax')(x)
model = Model(input_layer, output_layer)
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy']
             )
```

```python
Epochs=10
history = model.fit(train_data, epochs=Epochs, validation_data=val_data,verbose=2)
```

```python
def plot_model_history(history):
    plt.figure(figsize=(15,5))
    plt.subplot(1,2,1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    # Plot training and validation loss
    plt.subplot(1,2,2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
plot_model_history(history)
```

# Using VGG16 with pretrained weights

```python
base_model = tf.keras.applications.vgg16.VGG16(include_top=False,weights='imagenet', input_tensor=input_layer)
base_model.trainable = False
base_model.summary()
```

```python
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512,activation='relu')(x)
outputs = Dense(4, activation='softmax')(x)
vgg_model = Model(input_layer, outputs)
vgg_model.summary()
```

```python
vgg_model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy']
             )
```

```python
Epochs=10
history2 = vgg_model.fit(train_data, epochs=Epochs, validation_data=val_data,verbose=2)
```

```python
plot_model_history(history2)
```

# Deep Neural Network with Residual Block

```python
from tensorflow.keras.layers import Dense, MaxPooling2D, Conv2D, Dropout, BatchNormalization, Input, Flatten, GlobalAveragePooling2D, Concatenate,Add
from tensorflow.keras.models import Model

input_layer = Input(shape=(224,224,3))
x = Conv2D(32, 3, activation = 'relu',padding = 'same')(input_layer)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x_con1 = MaxPooling2D(2,2)(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x_con1)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = Add()([x,x_con1])
x_con2 = MaxPooling2D(2,2)(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x_con2)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = Add()([x,x_con2])
x_con3 = MaxPooling2D(2,2)(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x_con3)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = Add()([x,x_con3])
x_con4 = MaxPooling2D(2,2)(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x_con4)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = Conv2D(64, 3, activation = 'relu', padding = 'same')(x)
x = Add()([x,x_con4])
x = MaxPooling2D(2,2)(x)
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation = 'relu')(x)
x = Dropout(0.3)(x)
output_layer = Dense(4, activation = 'softmax')(x)
model3 = Model(input_layer, output_layer)
model3.summary()
```

```python
model3.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy']
             )
```

```python
Epochs=10
history3 = model3.fit(train_data, epochs=Epochs, validation_data=val_data,verbose=2)
```

```python
def plot_model_history(history):
    plt.figure(figsize=(15,5))
    plt.subplot(1,2,1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    # Plot training and validation loss
    plt.subplot(1,2,2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
```

```python
plot_model_history(history3)
```
